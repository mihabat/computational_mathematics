```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
```

```python
l = np.array([2, 1, 1/2, 1/4, 1/8, 1/16])
```

```python
mistake = np.array([0.2134561193240350, 0.0144206659922137, 0.0013392754437005, 0.0001446945317061, 0.0000168266944451, 0.0000020290933938])
mistake_cheb = np.array([0.1534957869823721, 0.0098655485704335, 0.0008929725633831, 0.0000952263686422, 0.0000110015346355, 0.0000013222917254])

fig, ax = plt.subplots(figsize=(10, 10), dpi=55)
ax.grid()
for i in range(len(l)):
    plt.errorbar(l[i], mistake[i], color='r',   fmt='.', markersize = 10)
for i in range(len(l)):
    plt.errorbar(l[i], mistake_cheb[i], color='b',   fmt='.',  markersize = 10)
ax.text(1.0, 0.2, r'для равномерных узлов', color = 'r', fontsize=15)
ax.text(1.0, 0.193, r'для Чебышевских узлов', color = 'b', fontsize=15)
plt.title("График зависимости max ошибки интерполяции от длины отрезка для N = 3")
plt.xlabel("длина отрезка")
plt.ylabel("max ошибка интерполяции")


fig, ax = plt.subplots(figsize=(10, 10), dpi=55)
ax.grid()
for i in range(len(l)):
    plt.errorbar(np.log(l[i]), np.log(mistake[i]), color='r',   fmt='.', markersize = 15)
for i in range(len(l)):
    plt.errorbar(np.log(l[i]), np.log(mistake_cheb[i]), color='b',   fmt='.', markersize = 15)
ax.text(-2.0, -4, r'для равномерных узлов', color = 'r', fontsize=15)
ax.text(-2.0, -4.4, r'для Чебышевских узлов', color = 'b', fontsize=15)
plt.title("График зависимости max логарифма ошибки интерполяции от логарифма длины отрезка для N = 3")
plt.xlabel("логарифм длины отрезка")
plt.ylabel("max логарифм ошибки интерполяции")
plt.show()
```
