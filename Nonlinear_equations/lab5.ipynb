```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
```

```python
y1 = []
x1 = np.linspace(1, 50, 50)
E1 = 0.861264884868175437
with open("e1.txt", 'r') as f:
    for line in f:
        y1.append(*[float(x) for x in line.split()])
y1 = np.array(y1)
y1 = np.log10(np.abs(y1 - E1))
```

```python
y2 = []
x2 = np.linspace(1, 50, 50)
E2 = 0.9478282237995902
with open("e2.txt", 'r') as f:
    for line in f:
        y2.append(*[float(x) for x in line.split()])
y2 = np.array(y2)
y2 = np.log10(np.abs(y2 - E2))
```

```python
y3 = []
x3 = np.linspace(1, 50, 50)
E3 = 1.26170305525310177130
with open("e3.txt", 'r') as f:
    for line in f:
        y3.append(*[float(x) for x in line.split()])
y3 = np.array(y3)
y3 = np.log10(np.abs(y3 - E3))
```

```python
y4 = []
x4 = np.linspace(1, 50, 50)
E4 = 1.5853138613542082165987
with open("e4.txt", 'r') as f:
    for line in f:
        y4.append(*[float(x) for x in line.split()])
y4 = np.array(y4)
y4 = np.log10(np.abs(y4 - E4))
```

```python
fig, ax = plt.subplots(figsize=(10, 10), dpi=55)
ax.grid()
plt.plot(x1, y1, label="e=0.1")
plt.plot(x2, y2, label="e=0.2")
plt.plot(x3, y3, label="e=0.5")
plt.plot(x4, y4, label="e=0.8")
plt.legend(prop={'size': 20})
ax.set_xlim(1, 10)
plt.title("График зависимости десятичного логарифма ошибки от номера итерации")
plt.xlabel("число итераций")
plt.ylabel("десятичный логарифм ошибки")
```
